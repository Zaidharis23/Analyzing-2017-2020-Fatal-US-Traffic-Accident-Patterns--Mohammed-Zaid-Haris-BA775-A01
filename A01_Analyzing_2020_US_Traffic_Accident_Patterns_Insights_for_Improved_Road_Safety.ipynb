**Key Findings**<br>
*  Demographic factors
	1. The high population States of California, Florida, Texas, New York, and Georgia report the most accidents of any states. California is at the top with 4,881 fatal accidents from 2017-2020
	2. The states of North Dakota, Vermont, and Wyoming have the fewest accidents at often less than 50 per year. This is relfective of their lower populations and traffic density
	3. The rural states of Wyoming and Montana face elevated fatal accident rates relative to their populations (Despite having low absolute accident counts) due to factors such as limited emergency services, harsh weather, and extensive rural road networks
	4. The urbanized states of New York and Massachusetts have low fatal accident rates relative to their populations due to having safer roads through extensive public transportation, better infrastructure, and effective urban design.
* Date and time
	1. An interactive tool was created in Tableau that allows stakeholders to understand when fatal crashes most frequently happen historically by year (2017- 2020 individually and combined), day, hour, and 10-minute block within that hour. For example, “On Sundays from 2017-2020, crashes between 12:00 am and 1:00 am occurred most frequently between 12:00 am and 12:10 am”
	2. Higher crash counts in 2020 compared to 2017-2019 (+5.06%) suggest that abnormal situations related to a greater number of drivers on the road lead to more crashes
	3. Fatal crashes follow a highly predictable pattern across years, where they are at their peak in the warmer months (June-October) and decrease from there to their lowest numbers in the middle of winter (January & February) before increasing again
* Relation Between Crashes and Driving Conditions
	1. Roadway departures, vehicle rollovers, and motor vehicle collisions are consistently the top causes of crashes from 2017-2020. These trends are often amplified by driver impairment due to alcohol or drugs
	2. Weekends account for the majority of drunk driving deaths, where Saturday has the most, followed by Sunday and Friday. This trend aligns with peak social drinking nights, and overall death numbers are in line with increased drunk driving accidents
*  Crash involvement characteristics (Vehicle and collision type)
	1. Parked or stopped vehicles contribute significantly to fatal crashes, and this number has increased over time. This challenges the assumption that most crashes occur between moving vehicles
* Weather conditions
	1.  Cloudy and rainy weather conditions contribute the most to crashes in both urban and rural settings
	2. Less frequent conditions such as fog and smog contribute to crashes at a higher rate in urban settings
	3. Accidents occur more frequently in the daylight hours due to a higher traffic volume, while many nighttime crashes occur in poorly lit rural areas
* People involved in crashes
	1. Most traffic fatalities occur within the 45-60 age group
	2. The under-18 age group experiences the smallest number of traffic fatalities
	3. Males account for the majority of fatalities across all age groups

**Joining the tables: accident, vevent & drimpair tables for the years 2017-2020**<br>
We joined the relevant tables to collect the data about the cause of accident and driver conditions at the time of crash together over the years.

```python
%%bigquery
CREATE OR REPLACE TABLE ba775-team-project-440113.ba775_project_join_tables.accident_action_combined AS
(WITH top_2020 AS (
  SELECT
    sequence_of_events_name AS Cause_of_Accident,
    COUNT(DISTINCT consecutive_number) AS Number_of_Accidents,
    '2020' AS Year
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2020` A
    LEFT JOIN `bigquery-public-data.nhtsa_traffic_fatalities. vevent_2020` V
    USING(consecutive_number)
    LEFT JOIN `bigquery-public-data.nhtsa_traffic_fatalities. drimpair_2020` D
    USING(consecutive_number)
  WHERE condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Not Reported%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%None/%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Unknown%'
  GROUP BY sequence_of_events_name
  ORDER BY Number_of_Accidents DESC
),
top_2019 AS (
  SELECT
    sequence_of_events_name AS Cause_of_Accident,
    COUNT(DISTINCT consecutive_number) AS Number_of_Accidents,
    '2019' AS Year
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2019` A
    LEFT JOIN `bigquery-public-data.nhtsa_traffic_fatalities. vevent_2019` V
    USING(consecutive_number)
    LEFT JOIN `bigquery-public-data.nhtsa_traffic_fatalities. drimpair_2019` D
    USING(consecutive_number)
  WHERE condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Not Reported%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%None/%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Unknown%'
  GROUP BY sequence_of_events_name
  ORDER BY Number_of_Accidents DESC
),
top_2018 AS (
  SELECT
    sequence_of_events_name AS Cause_of_Accident,
    COUNT(DISTINCT consecutive_number) AS Number_of_Accidents,
    '2018' AS Year
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2018` A
    LEFT JOIN `bigquery-public-data.nhtsa_traffic_fatalities. vevent_2018` V
    USING(consecutive_number)
    LEFT JOIN `bigquery-public-data.nhtsa_traffic_fatalities. drimpair_2018` D
    USING(consecutive_number)
  WHERE condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Not Reported%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%None/%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Unknown%'
  GROUP BY sequence_of_events_name
  ORDER BY Number_of_Accidents DESC
),
top_2017 AS (
  SELECT
    sequence_of_events_name AS Cause_of_Accident,
    COUNT(DISTINCT consecutive_number) AS Number_of_Accidents,
    '2017' AS Year
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2017` A
    LEFT JOIN `bigquery-public-data.nhtsa_traffic_fatalities. vevent_2017` V
    USING(consecutive_number)
    LEFT JOIN `bigquery-public-data.nhtsa_traffic_fatalities. drimpair_2017` D
    USING(consecutive_number)
  WHERE condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Not Reported%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%None/%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Unknown%'
  GROUP BY sequence_of_events_name
  ORDER BY Number_of_Accidents DESC
)
SELECT
  Year,Cause_of_Accident,
  SUM(CASE WHEN Year = '2020' THEN Number_of_Accidents ELSE 0 END) AS Number_of_Accidents_2020,
  SUM(CASE WHEN Year = '2019' THEN Number_of_Accidents ELSE 0 END) AS Number_of_Accidents_2019,
  SUM(CASE WHEN Year = '2018' THEN Number_of_Accidents ELSE 0 END) AS Number_of_Accidents_2018,
  SUM(CASE WHEN Year = '2017' THEN Number_of_Accidents ELSE 0 END) AS Number_of_Accidents_2017,
  MAX(SUM(Number_of_Accidents)) OVER (PARTITION BY Cause_of_Accident) AS Total_Number_of_Accidents
FROM (
  SELECT * FROM top_2020
  UNION ALL
  SELECT * FROM top_2019
  UNION ALL
  SELECT * FROM top_2018
  UNION ALL
  SELECT * FROM top_2017
) AS combined
GROUP BY Cause_of_Accident,Year
ORDER BY Total_Number_of_Accidents DESC
);

```

**Joining the tables: accident 2017-2020 & population**<br>
We joined the relevant tables to collect the data about the total number crashes, drunk drivers, fatalities and the population for each state.  

```python
%%bigquery
CREATE OR REPLACE TABLE `ba775-team-project-440113.ba775_project_join_tables.state_accident_population` AS
WITH state_accident_count AS (
  SELECT * FROM
  (
    SELECT
      '2020' AS Year,
      state_name,
      SUM(accident.number_of_fatalities) AS total_fatalities,
      SUM(accident.number_of_drunk_drivers) AS total_drunk_drivers,
      COUNT(*) AS total_accidents
    FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2020` AS accident
    GROUP BY state_name, Year

    UNION ALL

    SELECT
      '2019' AS Year,
      state_name,
      SUM(accident.number_of_fatalities) AS total_fatalities,
      SUM(accident.number_of_drunk_drivers) AS total_drunk_drivers,
      COUNT(*) AS total_accidents
    FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2019` AS accident
    GROUP BY state_name, Year

    UNION ALL

    SELECT
      '2018' AS Year,
      state_name,
      SUM(accident.number_of_fatalities) AS total_fatalities,
      SUM(accident.number_of_drunk_drivers) AS total_drunk_drivers,
      COUNT(*) AS total_accidents
    FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2018` AS accident
    GROUP BY state_name, Year

    UNION ALL

    SELECT
      '2017' AS Year,
      state_name,
      SUM(accident.number_of_fatalities) AS total_fatalities,
      SUM(accident.number_of_drunk_drivers) AS total_drunk_drivers,
      COUNT(*) AS total_accidents
    FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2017` AS accident
    GROUP BY state_name, Year
  )
)
SELECT
  a.Year,
  a.state_name,
  a.total_fatalities,
  a.total_drunk_drivers,
  a.total_accidents,
  p.Population AS population
FROM
  state_accident_count AS a
LEFT JOIN
  `ba775-team-project-440113.state_population_2017_2020.population` AS p
ON
  LOWER(a.state_name) = LOWER(p.State) AND a.Year = CAST(p.Year AS STRING);
```

### Analysis of Crashes with Relation to Day and Time

**<font size="3">Most Frequent 10-minute Crash Window by Hour by Day (2017-2020)</font>**

**Refer to Tableau Dashboard 1: Top Crash Times by Day & Hour(2017-2020)**

```python
%%bigquery
WITH combined_data AS (
  SELECT
    year,
    hour_of_crash,
    minute_of_crash,
    day_of_week_name
  FROM (
    SELECT
      2017 AS year,
      hour_of_crash,
      minute_of_crash,
      day_of_week_name
    FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2017`
    UNION ALL
    SELECT
      2018 AS year,
      hour_of_crash,
      minute_of_crash,
      day_of_week_name
    FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2018`
    UNION ALL
    SELECT
      2019 AS year,
      hour_of_crash,
      minute_of_crash,
      day_of_week_name
    FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2019`
    UNION ALL
    SELECT
      2020 AS year,
      hour_of_crash,
      minute_of_crash,
      day_of_week_name
    FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2020`
  )
),
ten_minute_windows AS (
  SELECT
    year,
    day_of_week_name,
    hour_of_crash,
    CASE
      WHEN minute_of_crash >= 0 AND minute_of_crash < 10 THEN '0-10'
      WHEN minute_of_crash >= 10 AND minute_of_crash < 20 THEN '10-20'
      WHEN minute_of_crash >= 20 AND minute_of_crash < 30 THEN '20-30'
      WHEN minute_of_crash >= 30 AND minute_of_crash < 40 THEN '30-40'
      WHEN minute_of_crash >= 40 AND minute_of_crash < 50 THEN '40-50'
      WHEN minute_of_crash >= 50 AND minute_of_crash < 60 THEN '50-60'
    END AS ten_minute_window,
    COUNT(*) AS crash_count
  FROM
    combined_data
  GROUP BY
    year, day_of_week_name, hour_of_crash, ten_minute_window
),
ranked_ten_minute_windows AS (
  SELECT
    year,
    day_of_week_name,
    hour_of_crash,
    ten_minute_window,
    crash_count,
    ROW_NUMBER() OVER (PARTITION BY year, day_of_week_name, hour_of_crash ORDER BY crash_count DESC) AS row_num
  FROM
    ten_minute_windows
),
top_ten_minute_each_hour AS (
  SELECT
    year,
    day_of_week_name,
    hour_of_crash,
    ten_minute_window
  FROM
    ranked_ten_minute_windows
  WHERE row_num = 1
),
aggregate_ten_minute_windows AS (
  SELECT
    day_of_week_name,
    hour_of_crash,
    ten_minute_window,
    SUM(crash_count) AS total_crash_count
  FROM
    ten_minute_windows
  GROUP BY
    day_of_week_name, hour_of_crash, ten_minute_window
),
ranked_aggregate_ten_minute_windows AS (
  SELECT
    day_of_week_name,
    hour_of_crash,
    ten_minute_window,
    total_crash_count,
    ROW_NUMBER() OVER (PARTITION BY day_of_week_name, hour_of_crash ORDER BY total_crash_count DESC) AS row_num
  FROM
    aggregate_ten_minute_windows
),
aggregate_top_ten_minute_each_hour AS (
  SELECT
    NULL AS year, -- Null for aggregated results
    day_of_week_name,
    hour_of_crash,
    ten_minute_window
  FROM
    ranked_aggregate_ten_minute_windows
  WHERE row_num = 1
),
combined_results AS (
  SELECT * FROM top_ten_minute_each_hour
  UNION ALL
  SELECT * FROM aggregate_top_ten_minute_each_hour
)
SELECT
  COALESCE(CAST(year AS STRING), "'17-'20 Agg") AS year, -- Replace NULL with '17-'20 Agg
  hour_of_crash,
  day_of_week_name AS day,
  ten_minute_window AS bucket
FROM
  combined_results
ORDER BY
  year,
  CASE day_of_week_name
    WHEN 'Sunday' THEN 1
    WHEN 'Monday' THEN 2
    WHEN 'Tuesday' THEN 3
    WHEN 'Wednesday' THEN 4
    WHEN 'Thursday' THEN 5
    WHEN 'Friday' THEN 6
    WHEN 'Saturday' THEN 7
  END,
  hour_of_crash;
```

**Analysis**

Interpretation: "On Sundays from 2017-2020, crashes between 12:00 am and 1:00 am occurred most frequently between 12:00 am and 12:10 am."

The output of this query does not provide any “surprising” or “shocking” results but provides one of this dataset's most valuable applications. Our initial analysis found the top crash hour by day for only 2020 (i.e., “On Saturdays in 2020, the most prevalent hour for crashes was 11:00 pm”). While this finding alone could be seen as beneficial for emergency services to prepare for the “most dangerous” hours on the road, some light experimentation with SQL led us to realize we could not only easily scale this to all hours of the day but do it for all years (Individually and on an aggregate level) and even identify the 10-minute increment within that top hour when crashes occurred most frequently.

With this output, we aimed to put it into practical use by creating an interactive dashboard that can be filtered by day and year to identify when crashes are most likely to occur by hovering over a point for a given hour. The most useful application would come when filtering for *‘17-’20 Agg* as this derives insights from the most data and provides the most generalized output for a given day and time.

With this dashboard, the vision is for law enforcement to pull it up daily and get an idea of when crashes are most likely to happen so they can optimize their prevention and response efforts. Whether this means getting more officers on an area’s most crash-prone roads or making more EMS available during peak crash windows, a heightened degree of readiness can significantly reduce car crashes and prevent these crashes from turning fatal.

Given that this tool relies on only four years of data, it is not perfect or foolproof. However, continuously adding crash data over time will increase the statistical backing of these findings and make the tool more and more reliable.

**Refer to Tableau Dashboard 1: Monthly Crash Counts by Year(2017-2020)**

```python
%%bigquery
SELECT
  month_of_crash_name AS month,
  SUM(CASE WHEN year_of_crash = '2017' THEN 1 ELSE 0 END) AS crash_count_2017,
  SUM(CASE WHEN year_of_crash = '2018' THEN 1 ELSE 0 END) AS crash_count_2018,
  SUM(CASE WHEN year_of_crash = '2019' THEN 1 ELSE 0 END) AS crash_count_2019,
  SUM(CASE WHEN year_of_crash = '2020' THEN 1 ELSE 0 END) AS crash_count_2020
FROM (
  SELECT '2020' AS year_of_crash, month_of_crash_name
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2020`
  UNION ALL
  SELECT '2019' AS year_of_crash, month_of_crash_name
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2019`
  UNION ALL
  SELECT '2018' AS year_of_crash, month_of_crash_name
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2018`
  UNION ALL
  SELECT '2017' AS year_of_crash, month_of_crash_name
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. accident_2017`
)
GROUP BY
  month_of_crash_name
ORDER BY
  CASE month_of_crash_name
    WHEN 'January' THEN 1
    WHEN 'February' THEN 2
    WHEN 'March' THEN 3
    WHEN 'April' THEN 4
    WHEN 'May' THEN 5
    WHEN 'June' THEN 6
    WHEN 'July' THEN 7
    WHEN 'August' THEN 8
    WHEN 'September' THEN 9
    WHEN 'October' THEN 10
    WHEN 'November' THEN 11
    WHEN 'December' THEN 12
  END;
```

**Analysis**

When examining this output, it is clear that fatal crashes are at their lowest in the winter months of January and February before increasing steadily until July, plateauing until October, and then steadily decreasing from November on. While 2020’s peak period was more accentuated, it followed roughly the same trend as 2017-2019, indicating that these trends tend to hold, regardless of circumstance.

These results are not surprising, as it is logical that more crashes would occur in the warmer months: People travel more, students are not in school, and the population is generally more inclined to be mobile in pursuit of warm-weather activities unavailable in the winter months. However, the strong definition of these trends is what stands out. The increase in crashes in the summer months is not an anomaly but entirely predictable.

It is likely impossible to stop this trend, but measures should be taken to mitigate it. Whether that be an increased emergency services presence on the road, increased fines for traffic violations (Speeding, DUI, reckless driving, etc.), or increased education/training for new drivers, anything to increase crash response, deter lousy driving, and/or improve driving abilities among the population, would make a positive impact in decreasing fatal crashes.

```python
%%bigquery
SELECT
  Cause_of_Accident,
    SUM(Number_of_Accidents_2020) AS Number_of_Accidents_2020,
  SUM(Number_of_Accidents_2019) AS Number_of_Accidents_2019,
  SUM(Number_of_Accidents_2018) AS Number_of_Accidents_2018,
  SUM(Number_of_Accidents_2017) AS Number_of_Accidents_2017,
  SUM(Total_Number_of_Accidents) AS Total_Number_of_Accidents
FROM ba775-team-project-440113.ba775_project_join_tables.accident_action_combined
GROUP BY Cause_of_Accident
LIMIT 10;
```

```python
%%bigquery
WITH accidents_by_year AS (
  SELECT
    '2017' AS Year,
    condition_impairment_at_time_of_crash_driver_name AS Driver_Condition,
    COUNT(consecutive_number) AS Number_of_Accidents
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. drimpair_2017`
  WHERE condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Not Reported%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%None/%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Unknown%'
  GROUP BY condition_impairment_at_time_of_crash_driver_name

  UNION ALL

  SELECT
    '2018' AS Year,
    condition_impairment_at_time_of_crash_driver_name AS Driver_Condition,
    COUNT(consecutive_number) AS Number_of_Accidents
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. drimpair_2018`
  WHERE condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Not Reported%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%None/%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Unknown%'
  GROUP BY condition_impairment_at_time_of_crash_driver_name

  UNION ALL

  SELECT
    '2019' AS Year,
    condition_impairment_at_time_of_crash_driver_name AS Driver_Condition,
    COUNT(consecutive_number) AS Number_of_Accidents
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. drimpair_2019`
  WHERE condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Not Reported%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%None/%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Unknown%'
  GROUP BY condition_impairment_at_time_of_crash_driver_name

  UNION ALL

  SELECT
    '2020' AS Year,
    condition_impairment_at_time_of_crash_driver_name AS Driver_Condition,
    COUNT(consecutive_number) AS Number_of_Accidents
  FROM `bigquery-public-data.nhtsa_traffic_fatalities. drimpair_2020`
  WHERE condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Not Reported%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%None/%'
    AND condition_impairment_at_time_of_crash_driver_name NOT LIKE '%Unknown%'
  GROUP BY condition_impairment_at_time_of_crash_driver_name
)

SELECT
  Driver_Condition,
  SUM(CASE WHEN Year = '2017' THEN Number_of_Accidents ELSE 0 END) AS Number_of_Accidents_2017,
  SUM(CASE WHEN Year = '2018' THEN Number_of_Accidents ELSE 0 END) AS Number_of_Accidents_2018,
  SUM(CASE WHEN Year = '2019' THEN Number_of_Accidents ELSE 0 END) AS Number_of_Accidents_2019,
  SUM(CASE WHEN Year = '2020' THEN Number_of_Accidents ELSE 0 END) AS Number_of_Accidents_2020
FROM accidents_by_year
GROUP BY Driver_Condition
ORDER BY SUM(Number_of_Accidents) DESC
LIMIT 5;
```

**Analysis**

Three major causes of accidents—running off the roadway (both right and left), motor vehicle in-transport events, and vehicle rollovers or overturns—consistently accounted for a significant number of incidents over the years. For instance, "Ran Off Roadway - Right" accidents increased from 2,862 in 2019 to 3,362 in 2020, while similar patterns are seen in motor vehicle in-transport events and rollovers. Other notable causes include crossing the centerline and collisions with standing objects like trees. A commonality across these incidents is the frequent impairment of drivers due to alcohol, drugs, or medications, which significantly exacerbates these risks. This highlights the urgent need for stricter enforcement against drunk driving, awareness campaigns, and preventive measures. Additionally, enhancing road safety awareness and promoting vehicle stability technologies can help reduce rollover and lane departure incidents, creating safer roads for everyone.

```python
%%bigquery
SELECT '2020' AS Year,
       CASE
           WHEN CAST(age AS INT64) < 18 THEN 'Under 18'
           WHEN CAST(age AS INT64) BETWEEN 18 AND 25 THEN '18-25'
           WHEN CAST(age AS INT64) BETWEEN 26 AND 35 THEN '26-35'
           WHEN CAST(age AS INT64) BETWEEN 36 AND 45 THEN '36-45'
           WHEN CAST(age AS INT64) BETWEEN 46 AND 60 THEN '46-60'
           WHEN CAST(age AS INT64) > 60 THEN 'Over 60'
           ELSE 'Not Reported'
       END AS age_group,
       SUM(CASE WHEN sex_name = 'Male' THEN 1 ELSE 0 END) AS male_fatal_accidents,
       SUM(CASE WHEN sex_name = 'Female' THEN 1 ELSE 0 END) AS female_fatal_accidents,
       SUM(CASE WHEN sex_name = 'Not Reported' THEN 1 ELSE 0 END) AS preferNotToSay_fatal_accidents,
       SUM(CASE WHEN sex_name IN ('Male', 'Female', 'Not Reported') THEN 1 ELSE 0 END) AS total_fatal_accidents
FROM `bigquery-public-data.nhtsa_traffic_fatalities. pbtype_2020`
WHERE SAFE_CAST(age AS INT64) IS NOT NULL
GROUP BY age_group
UNION ALL
SELECT '2019' AS Year,
       CASE
           WHEN CAST(age AS INT64) < 18 THEN 'Under 18'
           WHEN CAST(age AS INT64) BETWEEN 18 AND 25 THEN '18-25'
           WHEN CAST(age AS INT64) BETWEEN 26 AND 35 THEN '26-35'
           WHEN CAST(age AS INT64) BETWEEN 36 AND 45 THEN '36-45'
           WHEN CAST(age AS INT64) BETWEEN 46 AND 60 THEN '46-60'
           WHEN CAST(age AS INT64) > 60 THEN 'Over 60'
           ELSE 'Not Reported'
       END AS age_group,
       SUM(CASE WHEN sex_name = 'Male' THEN 1 ELSE 0 END) AS male_fatal_accidents,
       SUM(CASE WHEN sex_name = 'Female' THEN 1 ELSE 0 END) AS female_fatal_accidents,
       SUM(CASE WHEN sex_name = 'Not Reported' THEN 1 ELSE 0 END) AS preferNotToSay_fatal_accidents,
       SUM(CASE WHEN sex_name IN ('Male', 'Female', 'Not Reported') THEN 1 ELSE 0 END) AS total_fatal_accidents
FROM `bigquery-public-data.nhtsa_traffic_fatalities. pbtype_2019`
WHERE SAFE_CAST(age AS INT64) IS NOT NULL
GROUP BY age_group
UNION ALL
SELECT '2018' AS Year,
       CASE
           WHEN CAST(age AS INT64) < 18 THEN 'Under 18'
           WHEN CAST(age AS INT64) BETWEEN 18 AND 25 THEN '18-25'
           WHEN CAST(age AS INT64) BETWEEN 26 AND 35 THEN '26-35'
           WHEN CAST(age AS INT64) BETWEEN 36 AND 45 THEN '36-45'
           WHEN CAST(age AS INT64) BETWEEN 46 AND 60 THEN '46-60'
           WHEN CAST(age AS INT64) > 60 THEN 'Over 60'
           ELSE 'Not Reported'
       END AS age_group,
       SUM(CASE WHEN sex_name = 'Male' THEN 1 ELSE 0 END) AS male_fatal_accidents,
       SUM(CASE WHEN sex_name = 'Female' THEN 1 ELSE 0 END) AS female_fatal_accidents,
       SUM(CASE WHEN sex_name = 'Not Reported' THEN 1 ELSE 0 END) AS preferNotToSay_fatal_accidents,
       SUM(CASE WHEN sex_name IN ('Male', 'Female', 'Not Reported') THEN 1 ELSE 0 END) AS total_fatal_accidents
FROM `bigquery-public-data.nhtsa_traffic_fatalities. pbtype_2018`
WHERE SAFE_CAST(age AS INT64) IS NOT NULL
GROUP BY age_group
UNION ALL
SELECT '2017' AS Year,
       CASE
           WHEN CAST(age AS INT64) < 18 THEN 'Under 18'
           WHEN CAST(age AS INT64) BETWEEN 18 AND 25 THEN '18-25'
           WHEN CAST(age AS INT64) BETWEEN 26 AND 35 THEN '26-35'
           WHEN CAST(age AS INT64) BETWEEN 36 AND 45 THEN '36-45'
           WHEN CAST(age AS INT64) BETWEEN 46 AND 60 THEN '46-60'
           WHEN CAST(age AS INT64) > 60 THEN 'Over 60'
           ELSE 'Not Reported'
       END AS age_group,
       SUM(CASE WHEN sex_name = 'Male' THEN 1 ELSE 0 END) AS male_fatal_accidents,
       SUM(CASE WHEN sex_name = 'Female' THEN 1 ELSE 0 END) AS female_fatal_accidents,
       SUM(CASE WHEN sex_name = 'Not Reported' THEN 1 ELSE 0 END) AS preferNotToSay_fatal_accidents,
       SUM(CASE WHEN sex_name IN ('Male', 'Female', 'Not Reported') THEN 1 ELSE 0 END) AS total_fatal_accidents
FROM `bigquery-public-data.nhtsa_traffic_fatalities. pbtype_2017`
WHERE SAFE_CAST(age AS INT64) IS NOT NULL
GROUP BY age_group
ORDER BY
    CASE age_group
        WHEN 'Under 18' THEN 1
        WHEN '18-25' THEN 2
        WHEN '26-35' THEN 3
        WHEN '36-45' THEN 4
        WHEN '46-60' THEN 5
        WHEN 'Over 60' THEN 6
        ELSE 7
    END;
```
